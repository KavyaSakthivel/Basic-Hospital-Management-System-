```python
def update_doctor(doctor_id, name=None, specialization=None, degree=None, type=None, status=None):
    try:
        conn = connect_db()
        cursor = conn.cursor()
        query = "UPDATE Doctor SET "
        updates = []
        values = []

        if name:
            updates.append("name = %s")
            values.append(name)
        if specialization:
            updates.append("specialization = %s")
            values.append(specialization)
        if degree:
            updates.append("degree = %s")
            values.append(degree)
        if type:
            updates.append("type = %s")
            values.append(type)
        if status:
            updates.append("status = %s")
            values.append(status)

        query += ", ".join(updates) + " WHERE doctor_id = %s"
        values.append(doctor_id)

        cursor.execute(query, tuple(values))
        conn.commit()
        cursor.close()
        print(f"Doctor {doctor_id} updated successfully.")
    except Exception as e:
        print(f"Error updating doctor: {e}")
    finally:
        conn.close()
```

```python
def get_doctor_appointments(doctor_id):
    try:
        conn = connect_db()
        cursor = conn.cursor()
        cursor.execute("""
            SELECT appointment_id, patient_id, appointment_date, appointment_time, status
            FROM Appointment
            WHERE doctor_id = %s AND appointment_date >= CURRENT_DATE;
        """, (doctor_id,))
        appointments = cursor.fetchall()
        cursor.close()
        return appointments
    except Exception as e:
        print(f"Error retrieving doctor's appointments: {e}")
    finally:
        conn.close()

# Example of use
doctor_appointments = get_doctor_appointments(1)  # Doctor with ID 1
for appointment in doctor_appointments:
    print(appointment)
```

```python
def get_appointments_between_dates(user_type, user_id, start_date, end_date):
    try:
        conn = connect_db()
        cursor = conn.cursor()

        if user_type == "doctor":
            query = """
                SELECT appointment_id, patient_id, appointment_date, appointment_time, status
                FROM Appointment
                WHERE doctor_id = %s AND appointment_date BETWEEN %s AND %s;
            """
        elif user_type == "patient":
            query = """
                SELECT appointment_id, doctor_id, appointment_date, appointment_time, status
                FROM Appointment
                WHERE patient_id = %s AND appointment_date BETWEEN %s AND %s;
            """

        cursor.execute(query, (user_id, start_date, end_date))
        appointments = cursor.fetchall()
        cursor.close()
        return appointments
    except Exception as e:
        print(f"Error retrieving appointments between dates: {e}")
    finally:
        conn.close()

# Example usage:
from datetime import date

doctor_appointments = get_appointments_between_dates("doctor", 1, date(2024, 10, 1), date(2024, 12, 31))
for appointment in doctor_appointments:
    print(appointment)
```

```python
def patient_exists(patient_id):
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("SELECT 1 FROM patient WHERE patient_id = %s", (patient_id,))
    exists = cursor.fetchone() is not None
    cursor.close()
    conn.close()
    return exists
```

```python
def add_patient_book(patient_id):
    conn = connect_db()
    cursor = conn.cursor()
    
    # Prompt for all the required details
    name = input("Enter patient name: ")
    age = int(input("Enter patient age: "))
    gender = input("Enter patient gender (M/F): ")
    email = input("Enter patient email: ")
    phone = input("Enter patient phone number: ")
    admission_date = input("Enter admission date (YYYY-MM-DD): ")
    status = input("Enter patient status (e.g., Admitted/Discharged): ")

    # Insert the patient details into the patient table
    cursor.execute("""
        INSERT INTO patient (patient_id, name, age, gender, email, phone, admission_date, status)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
    """, (patient_id, name, age, gender, email, phone, admission_date, status))
    
    conn.commit()
    cursor.close()
    conn.close()
    print(f"Patient {patient_id} added to the database.")
```

```python
def book_appointment(patient_id, doctor_id, appointment_date, appointment_time):
    try:
        if not patient_exists(patient_id):
            # Generate a unique appointment_id, for example by using a timestamp or sequence from the database
            appointment_id = generate_appointment_id()  # Custom function to generate appointment_id
            add_patient_book(patient_id) 
        # Call the schedule_appointment function with the generated appointment_id
        schedule_appointment(
            appointment_id,
            appointment_date,
            appointment_time,
            doctor_id,
            patient_id,
            status="Pending"
        )
        print(f"Appointment booked successfully with ID: {appointment_id}")
    except Exception as e:
        print(f"Error booking appointment: {e}")

```

```python
def update_appointment(appointment_id, doctor_id=None, patient_id=None, appointment_date=None, appointment_time=None, status=None):
    conn = connect_db()
    cursor = conn.cursor()
    query = "UPDATE Appointment SET "
    params = []
    
    if doctor_id:
        query += "doctor_id = %s, "
        params.append(doctor_id)
    if patient_id:
        query += "patient_id = %s, "
        params.append(patient_id)
    if appointment_date:
        query += "appointment_date = %s, "
        params.append(appointment_date)
    if appointment_time:
        query += "appointment_time = %s, "
        params.append(appointment_time)
    if status:
        query += "status = %s "
        params.append(status)
    
    query += "WHERE id = %s;"
    params.append(appointment_id)

    cursor.execute(query, params)
    conn.commit()
    cursor.close()
    conn.close()
    print("Appointment updated successfully!")
```

```python
def rehash_user_passwords():
    conn = connect_db()
    cursor = conn.cursor()
    
    cursor.execute("""
        SELECT username, password FROM HosUser;
    """)
    
    users = cursor.fetchall()
    
    for username, password in users:
        # Check if password is not hashed (you might need to implement a check)
        if not password.startswith('$2b$'):  # bcrypt hashed passwords start with this prefix
            hashed_password = bcrypt.hashpw(password.encode('utf-8'), bcrypt.gensalt()).decode('utf-8')
            cursor.execute("""
                UPDATE HosUser SET password = %s WHERE username = %s;
            """, (hashed_password, username))
    
    conn.commit()
    cursor.close()
    conn.close()

rehash_user_passwords()
```

```python
def login_user(username, password):
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("SELECT password, role FROM HosUser WHERE username = %s;", (username,))
    user = cursor.fetchone()

    if user:
        stored_password, role = user
        if bcrypt.checkpw(password.encode('utf-8'), stored_password.encode('utf-8')):
            print(f"Login successful! Welcome, {role}.")
            return role
        else:
            print("Invalid password.")
    else:
        print("User not found.")

    cursor.close()
    conn.close()
    return None
```
